# Git: Comandos e Gitflow

## Principais Terminologias:
- **repositório**: lugar onde seus arquivos são salvos e rastreados, pode ser local ou remoto.
- **branch**: ramificação, podemos ser feito uma analogia à uma linha do tempo, existindo a principal (main) e as secundárias.
- **untracked**: arquivos novos que não foram inseridos no *git init* ou que ainda não foram colocados em *stage*.
- **stage**: um conceito aplicado a arquivos afetados pelo *git add*, semelhante a um palco, tudo que está em stage está apto a ser commitado.
- **commit/commitar**: ato de mandar arquivos para o repositório local, criando um savepoint, através do comando *git commit -m "mensagem"*.
- **modified**: arquivos que foram alterados depois de *commitados*.

## git init
- `git init`
- Inicia um repositório local `.git` no diretório atual.

## git clone
- `git clone https://github.com/owner/repo.git`
- Copia um repositório remoto para sua máquina, sincronizando branchs permitindo envios e recebimento de arquivos via CLI.

## git status
- `git status`
- Mostra arquivos modificados, não rastreados, e em *stage*.

## git add

- Adiciona arquivos não rastreados ao controle de versões do git.
- Após um commit, quaisquer alterações devem ser colocadas em *stage* com *git add* .

    - `git add .` (adiciona novos não rastreados e modificações, menos removidos)
    - `git add -u` (adiciona modificações e exclusões, menos arquivos novos não rastreados)
    - `git add -A` (novos não rastreados, modificações e removidos)

## git commit
- Envia ao repositório local um 'snapshot' do estado atual de seus arquivos, permitindo um ponto de salvamento.
- Git trabalha com algo semelhante a *ponteiros* para identificar o último commit, um commit altera o ponteiro para o novo arquivo versionado.

    - `git commit -m "mensagem"`
    - `git commit -a -m "Atualiza configurações"` (faz *git add* + *git commit*)
    - `git commit --amend -m "altera mensagem"` (edita a mensagem do ultimo commit)

## git pull
- Busca mudanças do repositório remoto e as mescla na sua branch atual local, atualizando sua branch com novas alterações.
- Só é possível dar *git pull* após ambos os repositórios estaremos "conectados".

    - `git pull origin main` (primeiro pull)
    - `git pull` (próximos pulls)

## git push
- Envia commits locais para um repositório remoto.

    - `git push origin main`
    - `git push -u origin minha-feature` (Enviar e definir upstream (-u), para futuros push/pull automáticos)

## git branch -m master main
- Muitas vezes ocorre de sua branch local se chamar 'master' e a remota 'main', logo é necessário renomear uma das branchs.
- Como a norma é usar 'main', renomeamos a 'master' -> 'main' com `git branch -m master main`
