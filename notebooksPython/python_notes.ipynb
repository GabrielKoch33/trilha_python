## Listas
Use lista quando:
- Você tem uma coleção ordenada de itens
- Geralmente do mesmo tipo, que: pode crescer, diminuir ou mudar
- Permite vários tipos
- Mantém ordem
- Permite repetição
- Permite alteração (mutável)
- Acessa por índice

```python
numeros = [10, 20, 30]
nomes = ["Ana", "Carlos", "João"]
misturado = [1, "A", True, [12, "3"]]
```

### Métodos de lista

```python
nomes.append('Gabriel')  # Adiciona "Gabriel" ao fim da lista
nomes.insert(1, "Luiz")  # Adiciona "Luiz" na posição [1], os elementos a direita são movidos [i+1]
numeros.remove(20)       # Remove a primeira ocorrência de um valor 20
nomes.pop()              # Remove o último valor da lista, ou pelo indice. Caso não exista o valor, gera erro.
nomes.pop(2)             # Remove o INDICE 2, não o valor 2
nomes.sort()             # Organiza a lista em ordem crescente
nomes.sort(reverse=True) # Organiza a lista em ordem decrescente
nomes.reverse()          # Inverte a ordem atual da lista. [3,2,6] -> [6,2,3]
nomes.count("Ana")       # Conta quantas vezes o nome Ana aparece na lista
nomes.index("Carlos")    # Retorna o indice do valor fornecido
```

## Sets
Use set quando você quer valores únicos e não se importa com a ordem.
- Não permite repetição
- Busca muito rápida
- Ótimo para remover duplicados
- Verificar existência rapidamente
- Encontrar interseção/diferença
- Não possuem índice
- Não garantem ordem
- Não aceitam duplicados
- Ótimos para operações de conjunto
- Aceita tipos mistos

### Métodos de sets

```python
s = {1, 2, 3}
s.add(7)          # Adiciona um valor
s.update([1, 2, 3])  # Adiciona VÁRIOS valores
s.discard(10)     # Caso não exista, não dá erro
s.clear()         # Esvazia o set

a = {1, 3, 5, 2, 6}
b = {3, 2, 5, 6, 8, 10}

a.union(b)                 # A + B, sem repetição
a.intersection(b)          # A ∩ B, existe em ambos
a.difference(b)            # A - B, exclusivo de A
a.symmetric_difference(b)  # (A - B) + (B - A), exclusivo de cada um
```

## List Comprehension
- É uma forma compacta de criar listas a partir de um loop
- **Quando usar:** quando se deseja criar uma lista a partir de outra
- **Quando não usar:** quando um `for` exige muitos `if`, acumuladores e outras variáveis

Forma comum:

```python
numeros = [1, 2, 3, 4, 5]

quadrados = []

for n in numeros:
    quadrados.append(n ** 2)

print(quadrados)
```

Com list comprehension:

```python
numeros = [1, 2, 3, 4, 5]

# nova_lista recebe [expressão sobre cada elemento de outra lista]
quadrados = [n ** 2 for n in numeros]

print(quadrados)
```

Basicamente:
Crie uma lista `quadrados`, onde cada elemento dela vai ser o resultado de percorrer outra lista potencializando seus valores `**2`.

Pode se usar com `if`:

```python
numeros = [1, 2, 3, 4, 5, 6]

pares = [numero for numero in numeros if numero % 2 == 0]
print(pares)

resultado = [x * 2 for x in numeros if x < 10 if x % 2 == 0]
            # (2)                  (1)
# 1 = Define um filtro para quais dados vão entrar na nova lista/são de nosso interesse
# 2 = Desses dados filtrados, insira eles com formatação XYZ, multiplique por ABC, tratamento de dados filtrados

resultado = [x * 2 if x > 2 else x for x in numeros]
# pergunta: "como esse elemento vai ser transformado?"
# [1, 2, 6, 8, 10] — todos entram, mas alguns são dobrados
```

## Set Comprehension

```python
nomes = ["Ana", "Bruno", "Ana", "Carlos", "Bruno"]

nomes_unicos = set()  # nomes_unicos vai aceitar apenas valores distintos

for nome in nomes:
    nomes_unicos.add(nome)

print(nomes_unicos)
```

## Dict Comprehension
- É uma forma compacta de criar dicionários a partir de um dicionário com valores percorridos de uma lista de forma compacta
- **Quando usar:** quando se deseja criar uma lista a partir de outra
- **Quando não usar:** quando um `for` exige muitos `if`, acumuladores e outras variáveis

Com dict normal:

```python
produtos = ["mouse", "teclado", "monitor"]
estoque = {}

for produto in produtos:
    estoque[produto] = 0

print(estoque)
```

Com comprehension:

```python
produtos = ["mouse", "teclado", "monitor"]

estoque = {produto: 0 for produto in produtos}
# adicione no dicionário 'estoque' como chave o valor que seja o produto atual em 'produtos', e atribua a ela o valor 0
print(estoque)
```

```python
precos = {
    "mouse": 80.0,
    "teclado": 150.0,
    "monitor": 900.0
}

precos_com_desconto = {produto: preco * 0.9 for produto, preco in precos.items()}
# dict precos_com_desconto recebe como chave o produto atual do dicionario precos e como valor, o preco * 0.9
# o .items() é para extrair tanto a chave como o valor de precos
print(precos_com_desconto)
```

## Generator Expression
Formas compactas de gerar valores; realizam *lazy evaluation*, recebendo valores um por um ao invés de acessar a lista inteira em memória.

- List comprehension usa `[]`, generator usa `()`
- Normalmente usado com: `sum()`, `any()`, `all()`, `max()`, `min()`
- Usasse *next()* para esvaziar um Generator

```python
total_vendido = sum(item[1] for item in 'vendas')          # soma de todos os campos de preços na tabela de vendas
tem_produto_caro = any(preco > 1000 for preco in 'precos')  # retorna True caso algum produto, pelo menos um, tenha preco > 1000
todos_positivos = all(numero > 0 for numero in 'numeros')    # retorna True se TODOS os numeros da lista forem positivos
```

Com list comprehension:

```python
valores = [item for item in 'vendas']
total = sum(valores)  # ao final disso sum() percorre toda a lista e faz += dos valores
```

Com generator expression:

```python
total = sum(item[1] for item in 'vendas')
# sum recebe os valores presentes em item[1] do valor atual em vendas, sem a necessidade de percorrer a
# lista inteira catando os valores; ao fim, total recebe o valor de sum()
```

### *args e **kwargs
É possível passar uma quantidade variada de parâmetros para uma função. Usar `*variavel` vai transformar a variável em uma TUPLA,
podendo receber quantos valores quiser. Isso ocorre pois números separados por vírgulas são interpretados como tuplas pelo Python.

```python
def soma(*args):  # * é o sinal de empacotamento num = ()
    total = 0
    for num in args:
        total += num
    return total


print(soma(2, 3, 4, 6))                # 15
print(soma(2, 3, 4, 6, 5, 5, 5, 1, 2))  # 32
print(soma(2, 3))                      # 5
```

`**kwargs` nos permite passar variáveis com palavras-chave como argumento. Kwargs são identificados com `var=valor`:

```python
def pessoa(**kwargs):  # basicamente um dicionário
    print(kwargs)
    for nome, idade in kwargs.items():
        print(f'{nome} tem atualmente {idade} anos de idade')


pessoa(gabriel='33', rafael='47', daniel='22')
# {'gabriel': '33', 'rafael': '47', 'daniel': '22'}
# gabriel tem atualmente 33 anos de idade
# rafael tem atualmente 47 anos de idade
# daniel tem atualmente 22 anos de idade
```

```python
def loja(nome, *produtos, **estoque):
    print(f'nome da loja: {nome}')

    for prod in produtos:
        print(prod)

    for chave, valor in estoque.items():
        print(f'Produto: {chave} | QTD: {valor}')


loja('bazar10', 'banana', 'bola', 'boneca', banana=10, bola=8, boneca=29)
```

### Lambda
Expressões lambda também são conhecidas como funções anônimas. Elas diferem das funções comuns por serem declaradas de maneira
diferente: sem definirmos nome para criá-las, utilizamos a palavra-chave `lambda`.

Temos que usar ou List Comprehension ou Gen Exp para combinar Lambdas = For Loops. 


Estrutura: `variavel = lambda argumento1, argumento2, argumento3: expressão`

```python
triplo = lambda x: x * 3
print(type(triplo))  # <class 'function'>
print(triplo(3))      # 9

quadrado = lambda x: x * x
print(quadrado(5))  # 25

# retorne x == True se o número for ímpar, se for par, retorne False
impar = lambda x: True if x % 2 == 1 else False
print(impar(1))  # True
print(impar(8))  # False

minha_lista = [('maçã', 2), ('banana', 1), ('laranja', 3)]
minha_lista_ordenada = sorted(minha_lista, key=lambda x: x[1])
    # por padrão o python ordenaria por string; para ordenar por valor, usamos key=
    # key= recebe o resultado de uma função aplicada sobre cada item, nesse caso os itens na posição [1]
    # sorted vai percorrer cada elemento de minha_lista e organizar em ordem em minha_lista_ordenada com base no campo [1] comparado entre itens
print(minha_lista_ordenada)

pessoas = [{'nome': 'Ana', 'idade': 25}, {'nome': 'Beto', 'idade': 22}]
ordenado = sorted(pessoas, key=lambda x: x['idade'])

a = lambda x, y: x * y
print(a(10, 5))

d = {
    'banana': 3,
    'apple': 5,
    'cherry': 1
}
# Sort by the second element (value) in each tuple
sorted_by_value = dict(sorted(d.items(), key=lambda item: item[1]))
# Result: {'cherry': 1, 'banana': 3, 'apple': 5}
```

## Datetime
O módulo `datetime` é utilizado para trabalhar com datas, horários e intervalos de tempo.

Importações mais comuns:

```python
from datetime import datetime
from datetime import date
from datetime import timedelta
```

### Data e hora atual

```python
agora = datetime.now()

# Retorna data e hora atuais do sistema.
# Exemplo:
# 2026-06-25 16:30:45.123456

print(agora)
```

### Data atual

```python
hoje = date.today()

# Retorna apenas a data atual.
# Exemplo:
# 2026-06-25

print(hoje)
```

### Acessando partes da data

```python
agora = datetime.now()

dia = agora.day
mes = agora.month
ano = agora.year

hora = agora.hour
minuto = agora.minute
segundo = agora.second

# Permite acessar componentes específicos da data/hora.

print(dia)
print(mes)
print(ano)
print(hora)
print(minuto)
print(segundo)
```

### Converter string para data

```python
texto_data = "25/06/2026"

data_convertida = datetime.strptime(
    texto_data,
    "%d/%m/%Y"
)

# strptime = String Parse Time
#
# Converte texto para objeto datetime.
#
# Formatações comuns:
#
# %d -> dia
# %m -> mês
# %Y -> ano com 4 dígitos
# %H -> hora
# %M -> minuto
# %S -> segundo

print(data_convertida)
```

### Converter data para string

```python
agora = datetime.now()

data_formatada = agora.strftime("%d/%m/%Y")

# strftime = String Format Time
#
# Converte um objeto datetime em texto.

print(data_formatada)
```

### Comparação de datas

```python
data1 = datetime(2026, 6, 1)
data2 = datetime(2026, 6, 25)

print(data2 > data1)
print(data2 < data1)
print(data2 == data1)

# True  -> data posterior
# False -> data anterior
```

### Diferença entre datas

```python
inicio = datetime(2026, 6, 1)
fim = datetime(2026, 6, 25)

diferenca = fim - inicio

# Retorna um objeto timedelta.

print(diferenca)
print(diferenca.days)

# Resultado:
# 24 days, 0:00:00
# 24
```

### Somar dias

```python
hoje = datetime.now()

amanha = hoje + timedelta(days=1)

# Adiciona dias à data.

print(amanha)
```

### Subtrair dias

```python
hoje = datetime.now()

ontem = hoje - timedelta(days=1)

# Remove dias da data.

print(ontem)
```

### Dia da semana

```python
data = datetime(2026, 6, 25)

dia_semana = data.weekday()

# Retorna:
#
# 0 -> Segunda
# 1 -> Terça
# 2 -> Quarta
# 3 -> Quinta
# 4 -> Sexta
# 5 -> Sábado
# 6 -> Domingo

print(dia_semana)
```

### Formatações mais utilizadas

```python
agora = datetime.now()

formato1 = agora.strftime("%d/%m/%Y")
# 25/06/2026

formato2 = agora.strftime("%d/%m/%Y %H:%M")
# 25/06/2026 16:30

formato3 = agora.strftime("%Y-%m-%d")
# 2026-06-25

formato4 = agora.strftime("%H:%M:%S")
# 16:30:45

print(formato1)
print(formato2)
print(formato3)
```
