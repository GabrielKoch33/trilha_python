## 1. Pilha (Stack)

**Conceito:** estrutura **LIFO** (*Last In, First Out* — o último que entra é o primeiro que sai).
Pense em uma pilha de pratos: você só consegue tirar o prato do topo.

**Operações principais:**
| Operação | O que faz | Complexidade |
|---|---|---|
| `push(item)` | insere no topo | O(1) |
| `pop()` | remove e retorna o topo | O(1) |
| `peek()` | olha o topo sem remover | O(1) |
| `is_empty()` | verifica se está vazia | O(1) |
| `size()` | quantidade de elementos | O(1) |

**Usos comuns:** desfazer/refazer (Ctrl+Z), verificação de parênteses balanceados, pilha de chamadas de função (call stack).


```python
class Pilha:
    """Implementação de uma Pilha (Stack) usando uma lista Python como armazenamento interno."""

    def __init__(self):
        self._itens = []  # lista interna que guarda os elementos

    def push(self, item):
        """Insere um item no topo da pilha. O(1)"""
        self._itens.append(item)

    def pop(self):
        """Remove e retorna o item do topo. O(1)"""
        if self.is_empty():
            raise IndexError("pop(): a pilha está vazia")
        return self._itens.pop()

    def peek(self):
        """Retorna o item do topo sem removê-lo. O(1)"""
        if self.is_empty():
            raise IndexError("peek(): a pilha está vazia")
        return self._itens[-1]

    def is_empty(self):
        """Retorna True se a pilha estiver vazia. O(1)"""
        return len(self._itens) == 0

    def size(self):
        """Retorna a quantidade de elementos na pilha. O(1)"""
        return len(self._itens)

    def __repr__(self):
        # topo à direita, para visualizar como uma pilha "de baixo para cima"
        return f"Pilha(base -> topo): {self._itens}"
```

```python
# --- Demonstração ---
pilha = Pilha()
pilha.push(1)
pilha.push(2)
pilha.push(3)
print(pilha)                 # Pilha(base -> topo): [1, 2, 3]
print("Topo:", pilha.peek()) # 3
print("Removido:", pilha.pop())  # 3
print(pilha)                 # Pilha(base -> topo): [1, 2]
print("Vazia?", pilha.is_empty())
print("Tamanho:", pilha.size())
```

**Nota pythônica:** para uso real, a própria lista do Python (`list`) já se comporta como pilha
com `.append()` (push) e `.pop()` (pop) — ambos O(1) no final da lista. Não é necessário criar uma classe,
a menos que você queira restringir a interface (evitar que alguém acesse o meio da estrutura).


## 2. Fila (Queue)

**Conceito:** estrutura **FIFO** (*First In, First Out* — o primeiro que entra é o primeiro que sai).
Pense em uma fila de banco: quem chega primeiro é atendido primeiro.

**Operações principais:**
| Operação | O que faz | Complexidade |
|---|---|---|
| `enqueue(item)` | insere no final | O(1) |
| `dequeue()` | remove e retorna o do início | O(1)* |
| `peek()` | olha o início sem remover | O(1) |
| `is_empty()` | verifica se está vazia | O(1) |
| `size()` | quantidade de elementos | O(1) |

\* usando `list` pura, remover do início é O(n) porque todos os elementos são deslocados.
Por isso, abaixo implementamos com `collections.deque`, que é otimizado para isso — mas construída
"na mão" (com uma lista) primeiro ajuda a entender o problema.

**Usos comuns:** filas de impressão, processamento de tarefas (task queue), busca em largura (BFS).


```python
class FilaComLista:
    """Implementação 'raiz' de Fila usando lista pura.
    Serve para entender o conceito, mas dequeue() é O(n) — veja a versão otimizada logo abaixo.
    """

    def __init__(self):
        self._itens = []

    def enqueue(self, item):
        """Insere um item no final da fila. O(1)"""
        self._itens.append(item)

    def dequeue(self):
        """Remove e retorna o item do início da fila. O(n) nesta versão, pois desloca a lista."""
        if self.is_empty():
            raise IndexError("dequeue(): a fila está vazia")
        return self._itens.pop(0)

    def peek(self):
        """Retorna o item do início sem removê-lo. O(1)"""
        if self.is_empty():
            raise IndexError("peek(): a fila está vazia")
        return self._itens[0]

    def is_empty(self):
        return len(self._itens) == 0

    def size(self):
        return len(self._itens)

    def __repr__(self):
        return f"Fila(início -> fim): {self._itens}"
```

```python
# --- Demonstração ---
fila = FilaComLista()
fila.enqueue("A")
fila.enqueue("B")
fila.enqueue("C")
print(fila)                  # Fila(início -> fim): ['A', 'B', 'C']
print("Início:", fila.peek())     # A
print("Removido:", fila.dequeue())  # A
print(fila)                  # Fila(início -> fim): ['B', 'C']
```

**Nota pythônica:** o módulo `collections` oferece `deque` (*double-ended queue*), uma lista
duplamente encadeada otimizada para inserir/remover em **ambas as pontas** em O(1). Para filas reais,
use sempre `deque` em vez de `list`.


```python
from collections import deque

fila_otimizada = deque()
fila_otimizada.append("A")       # enqueue
fila_otimizada.append("B")
fila_otimizada.append("C")
print(fila_otimizada)            # deque(['A', 'B', 'C'])
print("Removido:", fila_otimizada.popleft())  # dequeue em O(1)
print(fila_otimizada)            # deque(['B', 'C'])
```

## 3. Lista Ligada (Linked List)

**Conceito:** sequência de elementos (*nós*) onde cada nó guarda um valor e uma referência (ponteiro)
para o próximo nó. Diferente da lista do Python (que é um array dinâmico), os elementos **não** ficam
em posições contíguas de memória.


\* O(n) porque, sem um ponteiro para o final (`tail`), é preciso percorrer a lista inteira.

**Usos comuns:** implementação de outras estruturas (pilhas, filas), quando você precisa inserir/remover
com frequência sem realocar um array inteiro.


```python
class No:
    """Um nó individual da lista ligada: guarda um valor e aponta para o próximo nó."""

    def __init__(self, valor):
        self.valor = valor
        self.proximo = None  # por padrão, não aponta para lugar nenhum


class ListaLigada:
    """Lista ligada simples (singly linked list)."""

    def __init__(self):
        self.cabeca = None  # primeiro nó da lista (head)

    def inserir_inicio(self, valor):
        """Insere um novo nó no início da lista. O(1)"""
        novo_no = No(valor)
        novo_no.proximo = self.cabeca
        self.cabeca = novo_no

    def inserir_fim(self, valor):
        """Insere um novo nó no final da lista. O(n)"""
        novo_no = No(valor)
        if self.cabeca is None:
            self.cabeca = novo_no
            return
        atual = self.cabeca
        while atual.proximo is not None:
            atual = atual.proximo
        atual.proximo = novo_no

    def remover(self, valor):
        """Remove a primeira ocorrência do valor. Retorna True se removeu, False se não achou. O(n)"""
        atual = self.cabeca
        anterior = None

        while atual is not None:
            if atual.valor == valor:
                if anterior is None:       # o nó a remover é a cabeça
                    self.cabeca = atual.proximo
                else:
                    anterior.proximo = atual.proximo
                return True
            anterior = atual
            atual = atual.proximo

        return False  # valor não encontrado

    def buscar(self, valor):
        """Retorna True se o valor existir na lista. O(n)"""
        atual = self.cabeca
        while atual is not None:
            if atual.valor == valor:
                return True
            atual = atual.proximo
        return False

    def __repr__(self):
        valores = []
        atual = self.cabeca
        while atual is not None:
            valores.append(str(atual.valor))
            atual = atual.proximo
        return " -> ".join(valores) + " -> None"
```

```python
# --- Demonstração ---
lista = ListaLigada()
lista.inserir_fim(10)
lista.inserir_fim(20)
lista.inserir_fim(30)
lista.inserir_inicio(5)
print(lista)                     # 5 -> 10 -> 20 -> 30 -> None

lista.remover(20)
print(lista)                     # 5 -> 10 -> 30 -> None

print("Contém 10?", lista.buscar(10))  # True
print("Contém 99?", lista.buscar(99))  # False
```

**Nota pythônica:** o Python não tem uma lista ligada nativa porque a `list` embutida (array
dinâmico) já cobre a maioria dos casos de uso com acesso O(1) por índice. Quando você realmente precisa
de inserção/remoção O(1) nas duas pontas, o pythônico é usar `collections.deque` (que internamente é
implementado como uma lista duplamente ligada de blocos).


## 4. Tabela Hash (Hash Table)

**Conceito:** estrutura que associa **chaves a valores**, usando uma *função hash* para calcular
em qual posição (bucket) de um array cada chave deve ficar. Isso permite busca, inserção e remoção
em tempo médio **O(1)**.

Quando duas chaves diferentes caem na mesma posição (*colisão*), uma das soluções mais comuns é
o **encadeamento separado** (*separate chaining*): cada posição do array guarda uma lista de pares
(chave, valor).

**Operações principais:**
| Operação | O que faz | Complexidade média | Complexidade pior caso |
|---|---|---|---|
| `inserir(chave, valor)` | insere/atualiza | O(1) | O(n) |
| `buscar(chave)` | retorna o valor | O(1) | O(n) |
| `remover(chave)` | remove o par | O(1) | O(n) |
| `contem(chave)` | verifica existência | O(1) | O(n) |

O pior caso O(n) acontece quando muitas chaves colidem no mesmo bucket (má função hash ou tabela pequena).

**Usos comuns:** é exatamente como o `dict` do Python funciona por baixo dos panos; caches, contagem de
frequência, índices de bancos de dados.


```python
class TabelaHash:
    """Tabela hash com tratamento de colisão por encadeamento separado (separate chaining)."""

    def __init__(self, capacidade=8):
        self.capacidade = capacidade
        # cada posição (bucket) começa como uma lista vazia de pares [chave, valor]
        self._buckets = [[] for _ in range(capacidade)]
        self._quantidade = 0

    def _hash(self, chave):
        """Converte a chave em um índice válido dentro da capacidade da tabela."""
        return hash(chave) % self.capacidade

    def inserir(self, chave, valor):
        """Insere um par chave-valor. Se a chave já existir, atualiza o valor. O(1) em média."""
        indice = self._hash(chave)
        bucket = self._buckets[indice]

        for par in bucket:
            if par[0] == chave:
                par[1] = valor  # chave já existe: atualiza o valor
                return

        bucket.append([chave, valor])  # chave nova: adiciona ao bucket
        self._quantidade += 1

    def buscar(self, chave):
        """Retorna o valor associado à chave. Lança KeyError se não existir. O(1) em média."""
        indice = self._hash(chave)
        bucket = self._buckets[indice]

        for par in bucket:
            if par[0] == chave:
                return par[1]

        raise KeyError(f"chave não encontrada: {chave!r}")

    def remover(self, chave):
        """Remove o par associado à chave. Retorna True se removeu, False se não existia."""
        indice = self._hash(chave)
        bucket = self._buckets[indice]

        for i, par in enumerate(bucket):
            if par[0] == chave:
                bucket.pop(i)
                self._quantidade -= 1
                return True

        return False

    def contem(self, chave):
        """Retorna True se a chave existir na tabela. O(1) em média."""
        indice = self._hash(chave)
        return any(par[0] == chave for par in self._buckets[indice])

    def __len__(self):
        return self._quantidade

    def __repr__(self):
        pares = []
        for bucket in self._buckets:
            for chave, valor in bucket:
                pares.append(f"{chave!r}: {valor!r}")
        return "{" + ", ".join(pares) + "}"
```

```python
# --- Demonstração ---
tabela = TabelaHash()
tabela.inserir("nome", "Gabriel")
tabela.inserir("curso", "Estrutura de Dados")
tabela.inserir("linguagem", "Python")

print(tabela)                          # {'nome': 'Gabriel', 'curso': 'Estrutura de Dados', 'linguagem': 'Python'}
print("Busca:", tabela.buscar("nome")) # Gabriel

tabela.inserir("nome", "Gabriel Silva")  # atualiza valor existente
print("Após atualizar:", tabela.buscar("nome"))

print("Contém 'curso'?", tabela.contem("curso"))  # True
tabela.remover("curso")
print("Após remover:", tabela)
print("Quantidade de itens:", len(tabela))
```

**Nota pythônica:** o `dict` nativo do Python **é** uma tabela hash — implementada em C, com
tratamento de colisão (por *open addressing*, diferente do encadeamento que fizemos aqui), redimensionamento
automático e otimizações de baixo nível. Na prática, você sempre usará `dict` (ou `set`, que é uma tabela
hash sem valores associados). Implementar a sua própria, como fizemos, serve para entender **por que**
o `dict` é O(1) em média.


## Resumo comparativo

| Estrutura | Ordem | Melhor uso | Equivalente pythônico |
|---|---|---|---|
| Pilha | LIFO | desfazer ações, parênteses balanceados | `list` (`.append()` / `.pop()`) |
| Fila | FIFO | processamento em ordem de chegada | `collections.deque` |
| Lista Ligada | sequencial, sem índice | inserção/remoção frequente nas pontas | `collections.deque` (internamente) |
| Tabela Hash | associativa (chave-valor) | busca rápida por chave | `dict` / `set` |
