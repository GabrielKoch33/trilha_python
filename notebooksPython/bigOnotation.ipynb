# Complexidade de Algoritmos e Notação Big O

A notação Big O é uma notação matemática utilizada na ciência da computação para classificar algoritmos de acordo com o **crescimento do tempo de execução** em relação ao **tamanho do input**.

 A notação não nos retorna diretamente o tempo exato ou aproximado para a conclusão de um algoritmo, mas nos ajuda a entender o quanto nosso programa pode crescer e o quão eficiente e otimizado ele é, ignorando detalhes de hardware ou de linguagens de programação.

 O Termo 'n' representa o **tamanho da entrada**


### Explicação:

O(1): Tempo Constante.

Independente do tamanho da entrada, a operação terá sempre o mesmo custo.

Ex:


```python
def pegar_primeiro(lista):
    return lista[0]
```

O(log n): Tempo Logarítmico.

A cada novo passo, o número de instruções decresce pela metade. Um exemplo clássico é a busca binária, onde em um array de 1.000.000 de elementos, o número máximo de operações é +- 20.

```python
arr = [1,3,5,6,8,10,22,34,56,66,78,99,110]

def busca_binaria(lista, alvo):

    inicio, fim = 0, len(lista) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2

        if lista[meio] == alvo:
            return (f'Valor encontrado na posição: {meio}')
        
        elif lista[meio] < alvo:
            inicio = meio + 1

        else:
            fim = meio - 1
    return 'Valor não encontrado!'
```


- O(n): Tempo Linear.

O número de operações/esforço cresce na mesma proporção da entrada, 1:1.
Um exemplo são For Loops simples que verificam cada posição do vetor.

No caso abaixo, podemos encontrar o elemento na primeira, na quadrigésima ou na N-nésima comparação, sendo no pior caso o mesmo tanto de elementos do vetor


```python
def encontra_num_escolhido(lista):
    num_escolhido = 7

    for numero in lista:
        if numero == num_escolhido:
            return numero
```


- O(n log n): Linearítmico.

Comum em algoritmos de ordenação eficientes (merge sort, quicksort, timesort(o 'sorted()' do python usa esse algoritmo)). **log n** surge da divisão do problema pela metade (o que é muito rápido), porém essa divisão ocorre em **n** elementos diferentes, sendo esse caso **log n** para cada um dos itens.



- O(n²): Tempo Quadrático.

Geralmente aparece em casos que temos laços aninhados percorrendo a mesma entrada. Isso emplica que para cada elemento do laço externo, o número de operações realizadas pelo laço externo é igual ao quadrado de **n**.

| Tamanho | Nº Operações |
|---|---|
|     1     |  1    |
|2|4|
|3|9|
|10|100|

O(n³) seria o equivalente a 3 laços aninhados.

```python
def tem_duplicado(lista):
    for i in range(len(lista)):
        for j in range(len(lista)):
            if i != j and lista[i] == lista[j]:
                return True
    return False
```


- O(2ⁿ): Tempo Exponencial.

Cada novo elemento na entrada **dobra** o trabalho. Muito comum em algoritmos recursivos mal otimizados.

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
```

- O(n!): Tempo Fatorial.

Fatorial significa multiplicar **1** até **n**, fazendo com que o algoritmo cresça de forma absurda.
Um exemplo é o do caixeiro-viajante, onde um vendedor precisa passar por **n** cidades, passando por cada uma exatamente **uma vez**, e no final voltar a seu ponto de origem. Qual seria rota mais curta? Impossível dizer, pois com **5 cidades** temos +- 24 rotas possívei, com 15 cidades, aproximadamente 87 bilhões de opções.
