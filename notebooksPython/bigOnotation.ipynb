# Complexidade de Algoritmos e Notação Big O

A notação Big O é uma notação matemática utilizada na ciência da computação para classificar algoritmos de acordo com o **crescimento do tempo de execução** em relação ao **tamanho do input**.

 A notação não nos retorna diretamente o tempo exato ou aproximado para a conclusão de um algoritmo, mas nos ajuda a entender o quanto nosso programa pode crescer e o quão eficiente e otimizado ele é, ignorando detalhes de hardware ou de linguagens de programação.

 O Termo 'n' representa o **tamanho da entrada**


### Explicação:

O(1): Tempo Constante.

Independente do tamanho da entrada, a operação terá sempre o mesmo custo.

Ex:


```python
def pegar_primeiro(lista):
    return lista[0]
```

O(log n): Tempo Logarítmico.

A cada novo passo, o número de instruções decresce pela metade. Um exemplo clássico é a busca binária, onde em um array de 1.000.000 de elementos, o número máximo de operações é +- 20.

```python
arr = [1,3,5,6,8,10,22,34,56,66,78,99,110]

def busca_binaria(lista, alvo):

    inicio, fim = 0, len(lista) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2

        if lista[meio] == alvo:
            return (f'Valor encontrado na posição: {meio}')
        
        elif lista[meio] < alvo:
            inicio = meio + 1

        else:
            fim = meio - 1
    return 'Valor não encontrado!'
```


- O(n): Tempo Linear.

O número de operações/esforço cresce na mesma proporção da entrada, 1:1.
Um exemplo são For Loops simples que verificam cada posição do vetor.

No caso abaixo, podemos encontrar o elemento na primeira, na quadrigésima ou na N-nésima comparação, sendo no pior caso o mesmo tanto de elementos do vetor


```python
def encontra_num_escolhido(lista):
    num_escolhido = 7

    for numero in lista:
        if numero == num_escolhido:
            return numero
```
